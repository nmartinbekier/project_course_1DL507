# Read from Delta Share
Note that the common use case is to run a script like this, or connect to the Delta Share through an integration *outside* Databricks. This script here

```
%pip install delta-sharing
import delta_sharing
```

```
profile_path = f"/home/ubuntu/trase_uu/share/config.share"
share_name = f"candyland"
schema_name = f"candyland"
table_name = f"cd_2017"

client = delta_sharing.SharingClient(profile_path)

client.list_all_tables()
```

```
# Loading the table as a pandas dataframe
candy_df = delta_sharing.load_as_pandas(f"{profile_path}#{share_name}.{schema_name}.{table_name}")
```
